## Reducting memory by changing data types

```python
college = pd.read_csv("data/college.csv")
different_cols = ["RELAFFIL", "SATMTMID", "CURROPER", "INSTNM", "STABBR"]
cols2 = college.loc[ : , different_cols]
cols2.head()
```

```python
cols2.dtypes
```

```python
original_mem = cols2.memory_usage(deep=True)
original_mem
```

```python
cols2["RELAFFIL"] = cols2["RELAFFIL"].astype(np.int8)
```

```python
cols2.dtypes
```

```python
college[different_cols].memory_usage(deep=True)
```

```python
cols2.memory_usage(deep=True)
```

```python
cols2.select_dtypes(include=["object"]).nunique()
```

```python
cols2["STABBR"] = cols2["STABBR"].astype("category")
cols2.dtypes
```

```python
new_mem = cols2.memory_usage(deep=True)
new_mem
```

```python
new_mem / original_mem
```

```python
college.loc[0, "CURROPER"] = 10000000
college.loc[0, "INSTNM"] = college.loc[0, "INSTNM"] + "a"
college[["CURROPER", "INSTNM"]].memory_usage(deep=True)
```

```python
tsla = fdr.DataReader("tsla", start="2017-1-1")
tsla.head(8)
```

```python
tsla_close = tsla["Close"]
```

```python
tsla_cummax = tsla_close.cummax()
tsla_cummax.head(8)
```

```python
tsla_trailing_stop = tsla_cummax.mul(.9)
tsla_trailing_stop.head(8)
```

```python
def set_trailing_loss(symbol, purchase_date, perc):
    close = fdr.DataReader(symbol, start=purchase_date)["Close"]
    return close.cummax().mul(perc)
```

```python
set_trailing_loss("msft", "2017-6-1", .85)
```
