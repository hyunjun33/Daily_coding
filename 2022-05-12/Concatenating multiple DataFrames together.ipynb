## Concatenating multiple DataFrames together

```python
import pandas as pd
import numpy as np
```

```python
stocks_2016 = pd.read_csv("../python_cookbook/data/stocks_2016.csv", index_col="Symbol")
stocks_2017 = pd.read_csv("../python_cookbook/data/stocks_2017.csv", index_col="Symbol")
stocks_2016
```

```python
stocks_2017
```

```python
# concat 함수는 리스트를 받아서 이어붙인다. 기본설정은 행방향으로 두 객체를 이어붙인다.
s_list = [stocks_2016, stocks_2017]
pd.concat(s_list)
```

```python
# keys 매개변수로 각각의 데이터프레임 객체에 레이블을 설정할 수 있고, names 매개변수로 이름을 달수도 있다.
pd.concat(s_list, keys=["2016", "2017"], names=["Year", "Symbol"])
```

```python
# axis 매개변수로 붙일 방향을 설정할 수 있고, 기본설정은 outer join 외부 조인이다.
pd.concat(s_list, names=["Year", "Symbol"], keys=["2016", "2017"], axis=1)
```

```python
# 내부 조인으로 결합하고 싶을 경우, join 매개변수에 'inner'을 할당한다.
pd.concat(s_list, join="inner", keys=["2016", "2017"], names=["Year", None], axis=1)
```

```python
# append() 메서드도 내부적으로 concat 함수를 호출해서 작업한다.
stocks_2016.append(stocks_2017)
```
