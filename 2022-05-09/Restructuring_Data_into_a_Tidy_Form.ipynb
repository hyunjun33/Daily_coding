### Stacking multiple groups of variables simultaneously

```python
movie = pd.read_csv("../python_cookbook/data/movie.csv")
actor = movie[["movie_title", "actor_1_name", "actor_2_name",
               "actor_3_name", "actor_1_facebook_likes", "actor_2_facebook_likes",
               "actor_3_facebook_likes"]]
actor.head()
```

```python
def change_col_name(col_name):
    col_name = col_name.replace("_name", "")
    if "facebook" in col_name:
        fb_idx = col_name.find("facebook")
        col_name = (col_name[ :5] + col_name[fb_idx-1: ] + col_name[5:fb_idx-1])
    return col_name
```

```python
actor2 = actor.rename(columns=change_col_name)
actor2
```

```python
stubs = ["actor", "actor_facebook_likes"]
actor2_tidy = pd.wide_to_long(actor2, stubnames=stubs, i=["movie_title"], j="actor_num",
                              sep="_")
actor2_tidy.groupby(["movie_title"]).get_group("Avatar")
```

### Inverting stacked data

```python
usecol_func = lambda x: "UGDS_" in x or x == "INSTNM"
college = pd.read_csv("../python_cookbook/data/college.csv",
                      index_col="INSTNM",
                      usecols=usecol_func)
college
```

```python
college_stacked = college.stack()
college_stacked
```

```python
college_stacked.unstack()
```

```python
college2 = pd.read_csv("../python_cookbook/data/college.csv",
                       usecols=usecol_func)
college2
```

```python
college_melted = college2.melt(id_vars="INSTNM",
                               var_name="Race",
                               value_name="Percentage")
college_melted
```

```python
melted_inv = college_melted.pivot(index="INSTNM",
                                  columns="Race",
                                  values="Percentage")
melted_inv
```

```python
college2_replication = (melted_inv
                           .loc[college2["INSTNM"], college2.columns[1:]]
                           .reset_index()
                       )
college2.equals(college2_replication)
```

```python
college.stack().unstack(0)
```

```python
college.transpose()
```

### Unstacking after a groupby aggregation

```python
employee = pd.read_csv("../python_cookbook/data/employee.csv")
(employee
    .groupby("RACE")
    ["BASE_SALARY"]
    .mean()
    .astype(int)
)
```

```python
(employee
    .groupby(["RACE", "GENDER"])
    ["BASE_SALARY"]
    .mean()
    .astype(int)
)
```

```python
(employee
    .groupby(["RACE", "GENDER"])
    ["BASE_SALARY"]
    .mean()
    .astype(int)
    .unstack("GENDER")
)
```

```python
(employee
    .groupby(["RACE", "GENDER"])
    ["BASE_SALARY"]
    .mean()
    .astype(int)
    .unstack("RACE")
)
```

```python
(employee
    .groupby(["RACE", "GENDER"])
    ["BASE_SALARY"]
    .agg(["mean", "max", "min"])
    .astype(int)
)
```

```python
(employee
    .groupby(["RACE", "GENDER"])
    ["BASE_SALARY"]
    .agg(["mean", "max", "min"])
    .astype(int)
    .unstack("GENDER")
)
```

```python
(employee
    .groupby(["RACE", "GENDER"])
    ["BASE_SALARY"]
    .agg(["mean", "max", "min"])
    .unstack("RACE")
    .unstack("GENDER")
)
```

### Replicating pivot_table with a groupby aggregation

```python
flights = pd.read_csv("../python_cookbook/data/flights.csv")
fpt = flights.pivot_table(index="AIRLINE",
                          columns="ORG_AIR",
                          values="CANCELLED",
                          aggfunc="sum",
                          fill_value=0)
fpt
```

```python
(flights
    .groupby(["AIRLINE", "ORG_AIR"])
    ["CANCELLED"]
    .sum()
)
```

```python
(flights
    .groupby(["AIRLINE", "ORG_AIR"])
    ["CANCELLED"]
    .sum()
    .unstack("ORG_AIR", fill_value=0)
)
```

```python
flights.pivot_table(index=["AIRLINE", "MONTH"],
                    columns=["ORG_AIR", "CANCELLED"],
                    values=["DEP_DELAY", "DIST"],
                    aggfunc=["sum", "mean"],
                    fill_value=0)
```

```python
(flights
    .groupby(["AIRLINE", "MONTH", "ORG_AIR", "CANCELLED"])
    [["DEP_DELAY", "DIST"]]
    .agg(["sum", "mean"])
    .unstack(["ORG_AIR", "CANCELLED"], fill_value=0)
    .swaplevel(0, 1, axis=1)
)
```

### Renaming axis levels for easy reshaping

```python
college = pd.read_csv("../python_cookbook/data/college.csv")
(college
    .groupby(["STABBR", "RELAFFIL"])
    [["UGDS", "SATMTMID"]]
    .agg(["size", "min", "max"])
)
```

```python
(college
    .groupby(["STABBR", "RELAFFIL"])
    [["UGDS", "SATMTMID"]]
    .agg(["size", "min", "max"])
    .rename_axis(["AGG_COLS", "AGG_FUNCS"], axis="columns")
    .rename_axis(["STAB_NAME", "REL_NAME"], axis="index")
)
```

```python
(college
    .groupby(["STABBR", "RELAFFIL"])
    [["UGDS", "SATMTMID"]]
    .agg(["size", "min", "max"])
    .rename_axis(["AGG_COLS", "AGG_FUNCS"], axis="columns")
    .stack("AGG_FUNCS")
)
```

```python
(college
    .groupby(["STABBR", "RELAFFIL"])
    [["UGDS", "SATMTMID"]]
    .agg(["size", "min", "max"])
    .rename_axis(["AGG_COLS", "AGG_FUNCS"], axis="columns")
    .stack("AGG_FUNCS")
    .swaplevel("AGG_FUNCS", "STABBR", axis="index")
)
```

```python
(college
    .groupby(["STABBR", "RELAFFIL"])
    [["UGDS", "SATMTMID"]]
    .agg(["size", "min", "max"])
    .rename_axis(["AGG_COLS", "AGG_FUNCS"], axis="columns")
    .stack("AGG_FUNCS")
    .swaplevel("AGG_FUNCS", "STABBR", axis="index")
    .sort_index(level="RELAFFIL", axis="index")
    .sort_index(level="AGG_COLS", axis="columns")
)
```

```python
(college
    .groupby(["STABBR", "RELAFFIL"])
    [["UGDS", "SATMTMID"]]
    .agg(["size", "min", "max"])
    .rename_axis(["AGG_COLS", "AGG_FUNCS"], axis="columns")
    .stack("AGG_FUNCS")
    .unstack(["RELAFFIL", "STABBR"])
)
```

```python
(college
    .groupby(["STABBR", "RELAFFIL"])
    [["UGDS", "SATMTMID"]]
    .agg(["size", "min", "max"])
    .rename_axis(["AGG_COLS", "AGG_FUNCS"], axis="columns")
    .stack(["AGG_FUNCS", "AGG_COLS"])
)
```

```python
(college
    .groupby(["STABBR", "RELAFFIL"])
    [["UGDS", "SATMTMID"]]
    .agg(["size", "min", "max"])
    .rename_axis(["AGG_COLS", "AGG_FUNCS"], axis="columns")
    .unstack(["STABBR", "RELAFFIL"])
)
```

```python
(college
    .groupby(["STABBR", "RELAFFIL"])
    [["UGDS", "SATMTMID"]]
    .agg(["size", "min", "max"])
    .rename_axis([None, None], axis="index")
    .rename_axis([None, None], axis="columns")
)
```
