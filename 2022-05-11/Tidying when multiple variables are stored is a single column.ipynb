## Tidying when multiple variables are stored is a single column

```python
import pandas as pd
import numpy as np
```

```python
inspections = pd.read_csv("../python_cookbook/data/restaurant_inspections.csv",
                          parse_dates=["Date"])
inspections
```

```python
inspections.pivot(index=["Name", "Date"],
                  columns="Info", values="Value")
```

```python
(inspections.set_index(["Name", "Date", "Info"])
            .unstack("Info")
            .reset_index(col_level=-1)
)
```

```python
(inspections
    .set_index(["Name", "Date", "Info"])
    .unstack("Info")
    .reset_index(col_level=-1)
    .droplevel(0, axis=1)
    .rename_axis(None, axis=1)
)
```

```python
def flatten(df_):
    df_.columns = df_.columns.droplevel(0).rename(None)
    return df_

(inspections
    .set_index(["Name", "Date", "Info"])
    .unstack("Info")
    .reset_index(col_level=-1)
    .pipe(flatten)
)
```

```python
(inspections
    .set_index(["Name", "Date", "Info"])
    .squeeze()
    .unstack("Info")
    .reset_index()
    .rename_axis(None, axis=1)
)
```

```python
(inspections
    .pivot_table(index=["Name", "Date"],
                 columns="Info",
                 values="Value",
                 aggfunc="first")
    .reset_index()
    .rename_axis(None, axis=1)
)
```
