## Tidying when variables are stored in column names and values

```python
import pandas as pd
import numpy as np
```

```python
sensors = pd.read_csv("../python_cookbook/data/sensors.csv")
sensors
```

```python
(sensors
    .melt(id_vars=["Group", "Property"],
          var_name="Year")
)
```

```python
(sensors
    .melt(id_vars=["Group", "Property"],
          var_name="Year")
    .pivot_table(index=["Group", "Year"],
                 columns="Property",
                 values="value")
    .reset_index()
    .rename_axis(None, axis=1)
)
```

```python
(sensors
    .set_index(["Group", "Property"])
    .rename_axis("Year", axis=1)
    .stack()
    .unstack("Property")
    .rename_axis(None, axis="columns")
    .reset_index()
)
```
